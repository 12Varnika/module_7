```python
#4. . Write a Python program using multithreading where one thread adds numbers to a list, and another
#thread removes numbers from the list. Implement a mechanism to avoid race conditions using
#threading.Lock.
```

```python
import threading
import time
import random
shared_list = []
lock = threading.Lock()

def add_numbers():
    while True:
        number = random.randint(1, 100) 
        with lock:
            shared_list.append(number)
            print(f"Added: {number}. List now: {shared_list}")
        time.sleep(random.uniform(0.1, 1.0))  

def remove_numbers():
    while True:
        with lock:
            if shared_list:
                removed_number = shared_list.pop(0)
                print(f"Removed: {removed_number}. List now: {shared_list}")
            else:
                print("List is empty, waiting for numbers to be added.")
        time.sleep(random.uniform(0.1, 1.0)) 

add_thread = threading.Thread(target=add_numbers)
remove_thread = threading.Thread(target=remove_numbers)
add_thread.start()
remove_thread.start()
add_thread.join()
remove_thread.join()
```

```python
#7. Create a program that uses a thread pool to calculate the factorial of numbers from 1 to 10 concurrently.
#Use concurrent.futures.ThreadPoolExecutor to manage the threads.
```

```python
from concurrent.futures import ThreadPoolExecutor, as_completed

def factorial(n):
    """Calculate the factorial of a number n."""
    if n == 0 or n == 1:
        return 1
    else:
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result

def main():
    numbers = list(range(1, 11))  
    results = {}

    # Create a ThreadPoolExecutor to manage the threads
    with ThreadPoolExecutor(max_workers=5) as executor:
        # Submit tasks to the executor
        futures = {executor.submit(factorial, num): num for num in numbers}
        
        # Iterate over the completed futures as they finish
        for future in as_completed(futures):
            num = futures[future]
            try:
                result = future.result()
                results[num] = result
                print(f"Factorial of {num} is {result}")
            except Exception as e:
                print(f"An error occurred for number {num}: {e}")

if __name__ == "__main__":
    main()
```

```python
#8. Create a Python program that uses multiprocessing.Pool to compute the square of numbers from 1 to 10 in
#parallel. Measure the time taken to perform this computation using a pool of different sizes (e.g., 2, 4, 8
#processes).
```

```python
from multiprocessing import Pool
import time

def square(x):
    """Returns the square of x."""
    return x * x

def compute_with_pool_size(pool_size, numbers):
    """Computes the square of numbers using a pool of the given size and measures the time taken."""
    start_time = time.time()
    
    with Pool(pool_size) as pool:
        results = pool.map(square, numbers)
    
    end_time = time.time()
    duration = end_time - start_time
    
    print(f"Pool Size: {pool_size}, Results: {results}, Time Taken: {duration:.4f} seconds")
    return duration

def main():
    numbers = list(range(1, 11))  # Numbers from 1 to 10
    pool_sizes = [2, 4, 8]  # Different pool sizes to test
    
    for size in pool_sizes:
        compute_with_pool_size(size, numbers)

if __name__ == "__main__":
    main()
```
